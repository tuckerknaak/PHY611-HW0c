```python
#Imports
import matplotlib.pyplot as plt
from numpy import cos, pi, linspace
```

```python
#Evaulating differences between non-linear and harmonic oscillators

#Constants
A = 1
m = 1
w0 = 1
e1 = 1
e2 = 0.1
e3 = 0.01  #epsilon term approaching true harmonic oscillator

#Time interval of oscillations
time = linspace(-6*pi, 6*pi, 1000)

#Creating lists of positions for non-linear and harmonic oscillator
nonlinear1 = []
nonlinear2 = []
nonlinear3 = []
harmonic = []

#Evaluating position functions of non-linear oscillator with different e terms and harmonic oscillator
for t in time:
    nl1 = (A - (3*e1*A**3/4*m*w0**2))*cos(w0*t) + (e1*A**3/36*m*w0**2)*cos(3*w0*t) + (13*e1*A**3/18*m*w0**2)
    nl2 = (A - (3*e2*A**3/4*m*w0**2))*cos(w0*t) + (e2*A**3/36*m*w0**2)*cos(3*w0*t) + (13*e2*A**3/18*m*w0**2)
    nl3 = (A - (3*e3*A**3/4*m*w0**2))*cos(w0*t) + (e3*A**3/36*m*w0**2)*cos(3*w0*t) + (13*e3*A**3/18*m*w0**2)
    h = A*cos(w0*t)
    nonlinear1.append(nl1)
    nonlinear2.append(nl2)
    nonlinear3.append(nl3)
    harmonic.append(h)
```

```python
#Plotting the positions of each oscillator vs time
plt.figure(figsize=(8,4), dpi=100)
plt.plot(time, nonlinear1, linestyle='dashed', color='limegreen', label='Non-Linear $\epsilon=1$')
plt.plot(time, nonlinear2, linestyle='dashed', color='dodgerblue', label='Non-Linear $\epsilon=0.1$')
plt.plot(time, nonlinear3, linestyle='dashed', color='red', label='Non-Linear $\epsilon=0.01$')
plt.plot(time, harmonic, linestyle='solid', color='black', label='Harmonic')
plt.grid(True, linestyle='dashed', color='darkgray')
plt.xticks([-6*pi,-3*pi,0,3*pi,6*pi], ['-6$\pi$','-3$\pi$','0','3$\pi$','6$\pi$'])
plt.yticks([-1,-0.5,0,0.5,1])
plt.xlabel('Time [s]')
plt.ylabel('Position [m]')
plt.title('Non-Linear vs Harmonic Oscillations')
plt.legend(prop={'size':12.5}, bbox_to_anchor=(1, 0.6), loc='upper left', ncol=1)
plt.show()
```
