```python
#Imports
import matplotlib.pyplot as plt
from numpy import cos, sin, pi, linspace
```

```python
#Calculating the first n terms of the Fourier Series for the forcing function F(t) = F_0*(1+wt)

#Function to calculate Fourier coefficient for each n
def b_n(n):
    b = 2 * F_0 * (-1)**(n+1) / n
    return b

#Constants
F_0 = 1
w = 1

#Creating lists for containing data points of forcing function and Fourier Series
forcing_function = []
fourier_series = []

#Defining the interval of the function
time = linspace(-pi/w, pi/w, 100)

#Evaluating forcing function and Fourier Series for each point in time
N = int(input("How many terms of the Fourier Series do you want to evaluate? "))
for t in time:
    forcing = F_0 * (1 + w*t)
    forcing_function.append(forcing)
    
    fourier = 0  #setting value of Fourier Series to zero for each time
    for n in range(1,N):
        fourier += (b_n(n) * sin(n*w*t))
    fourier_series.append(1+fourier)
```

```python
#Plotting the forcing function and Fourier Series
plt.figure(figsize=(6,4), dpi=100)
plt.plot(time, forcing_function, linestyle='solid', color='black', label='$F_0$(1+$\omega$t)')
plt.plot(time, fourier_series, linestyle='dashed', color='red', label='f(t) for n=10')
plt.grid(True, linestyle='dashed', color='darkgray')
plt.xlabel('Time [s]')
plt.ylabel('Force [N]')
plt.title('Fourier Series Representation of Forcing Function')
plt.legend(prop={'size':12.5}, frameon=False)
plt.rcParams['legend.loc'] = 'upper left'
plt.show()
```
